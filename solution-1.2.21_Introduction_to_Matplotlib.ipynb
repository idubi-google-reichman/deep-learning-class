# Introduction to Matplotlib

Welcome to this interactive tutorial on Matplotlib! Matplotlib is a powerful library for creating a wide variety of plots and visualizations in Python. In this tutorial, you'll learn the basics of plotting with Matplotlib, how to use colors effectively, and how to create histograms.


## Basic Plotting

Let's start with the basics of plotting in Matplotlib. We will create a simple line plot showing a sine wave.


```python
import matplotlib.pyplot as plt
import numpy as np

# Data for plotting
t = np.linspace(0, 2*np.pi, 100)
s = np.sin(t)

plt.figure(figsize=(8, 4))
plt.plot(t, s, label='Sine Wave')
plt.title('Simple Plot')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```

## Drawing with Colors

Using colors can significantly enhance the visual appeal and clarity of plots. Let's see how to apply colors to plots.


```python
# Data for plotting
t = np.linspace(0, 2*np.pi, 100)
s1 = np.sin(t)
s2 = np.sin(t + np.pi/4)
s3 = np.sin(t - np.pi/4)
s4 = np.sin(t + np.pi/2)  # Additional sine wave

# Plotting each sine wave with different colors
plt.figure(figsize=(10, 5))
#Each figure can hold many plots
plt.plot(t, s1, color='r', label='Sine Wave 1 (Red)')    # Red
plt.plot(t, s2, color='green', label='Sine Wave 2 (Green)')  # Green
plt.plot(t, s3, color=[0,0,1], label='Sine Wave 3 (Blue)')   # Blue
plt.plot(t, s4, color='#FEA510', label='Sine Wave 4 (Yellow)') # Yellow

plt.title('Sine Waves with Colors')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()

```

## Creating Histograms

Histograms are a type of bar plot that shows the frequency of numerical data points. Let's create a histogram to understand the distribution of a dataset.


```python
# Generate random data
data = np.random.normal(0, 1, 1000)

plt.figure(figsize=(8, 4))
plt.hist(data, bins=30, color='b')
plt.title('Histogram of Randomly Generated Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

## Creating Subplots

Using subplots in Matplotlib allows you to place multiple plots in a single figure. This is particularly useful for comparing different datasets or different aspects of the same data. Let's create a grid of subplots showing different mathematical functions.


```python
import matplotlib.pyplot as plt
import numpy as np

# Data for plotting
t = np.linspace(0, 2*np.pi, 100)
s1 = np.sin(t)
s2 = np.cos(t)
s3 = np.tan(t)
s4 = np.sinh(t)

# Create a 2x2 grid of subplots
plt.figure(figsize=(5, 5))  # Set figure size

# Subplot 1: Sine function
plt.subplot(2, 2, 1)  # (rows, columns, panel number)
plt.plot(t, s1, 'r')
plt.title('Sine Function')
plt.grid(True)

# Subplot 2: Cosine function
plt.subplot(2, 2, 2)
plt.plot(t, s2, 'g')
plt.title('Cosine Function')
plt.grid(True)

# Handling the tan function by limiting y-values to prevent extreme values from distorting the graph
plt.subplot(2, 2, 3)
plt.plot(t, s3, 'b')
plt.ylim(-10, 10)  # Limit y-axis to make the tan function plot readable
plt.title('Tangent Function')
plt.grid(True)

# Subplot 4: Hyperbolic Sine function
plt.subplot(2, 2, 4)
plt.plot(t, s4, 'm')  # 'm' for magenta
plt.title('Hyperbolic Sine Function')
plt.grid(True)

# Adjust layout to prevent overlap
plt.tight_layout()

# Show plot
plt.show()
```
